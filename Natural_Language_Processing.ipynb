To translate each word or sentence from English to Spanish, French and German 


```python
pip install googletrans==3.1.0a0
```

```python
from googletrans import Translator
import numpy as np 
import pandas as pd 
```

```python
translator = Translator()
```

```python
df = pd.read_csv('/content/English.csv')
df
```

```python
df['Spanish'] = df['English words/sentences'].apply(translator.translate, dest='es').apply(getattr, args=('text',))
df['French'] = df['English words/sentences'].apply(translator.translate, dest='fr').apply(getattr, args=('text',))
df['German'] = df['English words/sentences'].apply(translator.translate, dest='de').apply(getattr, args=('text',))
```
